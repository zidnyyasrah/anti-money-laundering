
## **5.6 Feature Encoding**

Disini saya menggunakan count encoder dari library category-encoders, alasannya adalah kita bisa lihat bahwa skema money laundering ini biasanya dilakukan dalam beberapa hari agar bisa selesai. Jadi pertanyaannya adalah bagaimana cara menyimpan informasi akun, dan bank tersebut tanpa harus menggunakan one hot encoder, dimana dengan tingginya kardinalitas pada data penggunaan onehot encoder akan membuat jumlah feature akan menjadi sangat amat banyak.

Jadi count encoder disini akan berguna karena encoder ini akan menghitung jumlah frekuensi kemunculan value didalam dataset. Misal "Akun A" melakukan transaksi sebanyak 20x didalam dataset. Maka seluruh kemunculan "Akun A" tersebut akan di encode sebagai angka 20. Namun tentu saja tetap ada drawbacknya disini, jika ada 2 akun seperti "Akun A" dan "Akun B" yang sama-sama melakukan transaksi sebanyak 20x maka kedua akun tersebut akan diencode sebagai 20.

Tapi jika kita mengesampingkan drawback tersebut, informasi yang diprovide oleh encoder ini akan sangat berguna dikarenakan jumlah frekuensi dari transaksi money laundering merupakan informasi yang berharga. Seperti yang kita ketahui, pada skema money laundering itu biasanya tidak selesai dalam 1 transaksi saja, jadi ada beberapa transaksi agar sekma tersebut bisa complete.

```python
# Menampung fitur yang tergolong kategorik kedalam list
categorical_features = X_train.select_dtypes(include=['object']).columns

# Membuat Pipeline untuk kategorik encoder
categorical_transformer = Pipeline(
    steps=[("encoder", CountEncoder())]
)
```

## **5.7 Column Transformer**

Disini saya akan membuat preprocessor yang terisi features numeric dan categoric serta scaler dan encodernya menggunakan Column Transformer.

```python
# Membuat Column transformer untuk melakukan scaling dan encoding dan menggabungkan kedua tipe data tersebut
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("ord", categorical_transformer, categorical_features)],
        remainder = 'passthrough' # Untuk melewatkan selain fitur yg di transform
)
```

# **6 Model Definition**

Disini saya akan definisikan 2 jenis model, yang pertama adalah unsupervised learning yaitu anomaly detection dan juga supervised learning menggunakan data yang sudah dioversampling.

* Untuk yang kasus pertama (unsupervised), saya tidak melakukan handling imbalance data disana, karena memang specialty dari anomaly detection adalah untuk mencari data "anomaly" dari data yang memang ada ketidakseimbangan didalamnya.

* sedangkan pada kasus (supervised), akan sangat susah jika ada imbalance data, oleh karena itu saya akan handling data imbalance menggunakan metode smote yang sudah didefinisikan sebelumnya.

## **6.1 Unsupervised Learning - Anomaly Detection**

Pada unsupervised learning, saya akan mencoba 2 algoritma baseline, yaitu Isolation forest dan Local Outlier Factor. Kedua algoritma ini merupakan algoritma yang baik dalam kasus anomaly detection.

```python
# Membuat Pipeline untuk Isolation Forest
pipe_IF = Pipeline([
    ('preprocessor', preprocessor),
    ('IF', IsolationForest(random_state=42, n_jobs=-1))
])

# Membuat Pipeline untuk Local Outlier Factor
pipe_LOF = Pipeline([
    ('preprocessor', preprocessor),
    ('LOF', LocalOutlierFactor(n_jobs=-1, novelty=True))
])
```

## **6.2 Supervised Learning**

Pada supervised learning, saya akan menggunakan 2 algoritma baseline juga, logistic regression dan random forest.

```python
# Membuat Pipeline untuk Logistic Regression
pipe_smote_LR = Pipeline([
    ('preprocessor', preprocessor),
    ('smote', smote),
    ('LR', LogisticRegression(random_state=42, n_jobs=-1))
])

# Membuat Pipeline untuk Random Forest
pipe_smote_RF = Pipeline([
    ('preprocessor', preprocessor),
    ('smote', smote),
    ('RF', RandomForestClassifier(random_state=42, n_jobs=-1))
])
```

# **7. Model Training**

```python
# Train Isolation Forest
pipe_IF.fit(X_train)
```

```python
# Train Local Outlier Factor
pipe_LOF.fit(X_train)
```

```python
# Train Local Logistic Regression
pipe_smote_LR.fit(X_train, y_train)
```

```python
# Train Random Forest
pipe_smote_RF.fit(X_train, y_train)
```

## **Hyperparameter Tuning**

Pada bagian ini saya akan mencoba bermain pada hyperparameter `max_depth` dari algoritma random forest ini, dikarenakan waktu training yang cukup lama, saya tidak bisa melakukan eksperimen terhadap hyperparameter lainnya.

```python
# Define hyperparameter yang akan diujikan
random_grid_rf = {'RF__max_depth': [2,3,4]}
random_grid_rf
```
